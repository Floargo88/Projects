````python
def recategorize(category):
    if category == 0:
        return 'neutral'
    if category == 1:
        return 'positive'
    if category == 2:
        return 'positive'
    if category == 3:
        return 'positive'
    if category == 4:
        return 'positive'
    if category == -1:
        return 'negative'
    if category == -2:
        return 'negative'
    if category == -3:
        return 'negative'
    if category == -4:
        return 'negative'
    if category == -5:
        return 'negative'
    else:
        return 'Other'
````

````python
complete['sentiment'] = complete['sent_score'].apply(recategorize)
````

````python
import re
from collections import Counter
````

````python
texts = complete['full_text'].values.tolist()
````

````python
total_words = Counter()
for text in texts:
    # making text lower case
    text = text.lower()
    # removing URLS 
    text = text.split(' ')
    newtext = []
    for item in text:
        if 'http' not in item:
            newtext.append(item)
    
    newtext = ' '.join(newtext)
        
    # splitting the text in words (tokens)
    tokens = re.findall(r"[\w']+|[.,!?;$@]", newtext)
    for token in tokens:
        total_words[token] += 1
    
````

````python
total_words.most_common(100)
````
